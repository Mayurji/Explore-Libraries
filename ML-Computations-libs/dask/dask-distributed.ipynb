### Dask Distributed

```python
from dask.distributed import Client
```

```python
client = Client()
```

```python
client
```

```python
client = Client('127.0.0.1:39337')
```

```python
client.scheduler
```

```python
def square(x):
    return x**2
```

```python
def negative(x):
    return -x
```

```python
sq = client.map(square, range(1000))
neg = client.map(negative, sq)
total = client.submit(sum, neg)
```

```python
total
```

```python
total.result()
```

```python
client.gather(sq)
```

```python
client.gather(neg)
```

```python
client.restart()
```
