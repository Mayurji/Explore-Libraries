## Inner Working Of Neural Network



### **Neuron / Node**

#### **It is the smallest unit and building block of neural network. It takes a set of inputs, performs some mathematical computations, and outputs a response.**

### **Inputs and Outputs**

#### **Input and outputs are numbers, either positive or negative. In this example, we have two inputs and with one output response.**

#### **There is no limit on the number of inputs a neuron/node can take.**



### **Weighted Sum**

#### **Neuron takes the input and multiples it with corresponding weights. And all inputs are summed, and a bias term is added to it.**

### **Weights and Bias**

#### **These are parameters of the neural network. These are adjustable during the training phase.**

### **Example**



## **Activation**

#### The second computation done by the neuron is applying the activation function on top of the weighted-sum result.



### Applying Linear Activation

#### **Linear activation simply outputs the same input it receives.**
