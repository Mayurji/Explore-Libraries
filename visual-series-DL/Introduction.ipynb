## When to use ML/DL?

#### **Three phases of prediction automation**

#### First, Human judgement for a prediction based on experience or data. 
#### Second, Based on human experience, creating rules to make predicton. 
#### Third, model gets the experience from the data, and uses it for prediction.



#### As the data and rules become more complex, it makes sense to use ML/DL.

#### When you compare ML and DL, deep learning is more modular & flexible making it extremely adaptable.
